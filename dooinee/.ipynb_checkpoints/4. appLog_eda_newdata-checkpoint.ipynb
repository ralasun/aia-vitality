### Filter 1. 최종 보험가입 partyid 제외 2. 리워드유효여부

```python
# Filter

# 1. 최종 보험가입 partyid 제외 
df = df.join(members_df['멤버십통합'], on='PartyId')
assert len(df) == df_len, 'Error! length change'
df = df[df['멤버십통합']=='N']
df_len = len(df)
df = df.drop(columns=['멤버십통합'])

# 2. 리워드유효여부
df = df.join(members_df['가입일자'], on='PartyId')
assert len(df) == df_len, 'Error! length change'

diff_days = (df['방문일시'] - df['가입일자']).dt.floor('d')
diff_days = diff_days.dt.days

rewardIn_lst = []
for diff, join_date in zip(diff_days, df['가입일자']):
    if pd.isnull(join_date):
        rewardIn_lst.append(join_date)
        continue
    if join_date < pd.Timestamp(2020, 11, 4):
        if diff <= 168:
            rewardIn_lst.append('Y')
        else:
            rewardIn_lst.append('N')
    else:
        if diff <= 84:
            rewardIn_lst.append('Y')
        else:
            rewardIn_lst.append('N')
df['리워드기간'] = rewardIn_lst
df = df.drop(columns='가입일자')
```

```python
# applog df에 class join
df['month'] = df['방문일시'].dt.to_period('M')
df = pd.merge(left=df, right=seg_df, on=['PartyId','month'], how='left', sort=False)
assert len(df) == df_len, 'Error! length change'
#df = df.drop(columns='month')
gc.collect()
df
```

### 1-1. 세션별 페이지수 (세션길이)

```python
pagecode_df['카테고리1'].unique()
```

```python
# 세션별 페이지수 (세션 길이)


# groupby class
users_df = df[['PartyId', 'month', 'class']].groupby(['PartyId','month']).first()
users_len = len(users_df)

# 전체 페이지수
tmp_df = df[['PartyId', 'month', '방문일시']].groupby(['PartyId','month']).count()
tmp_df.columns = ['전체페이지수']
users_df = pd.merge(users_df, tmp_df, left_index=True, right_index=True)
assert len(users_df) == users_len, 'Error! length change'

# 전체 세션수
tmp_df = df[['PartyId', 'month','세션ID']].groupby(['PartyId','month'])['세션ID'].nunique().to_frame()
tmp_df.columns = ['전체세션수']
users_df = pd.merge(users_df, tmp_df, left_index=True, right_index=True)
assert len(users_df) == users_len, 'Error! length change'

# 세션별 페이지수
users_df['세션당페이지수'] = np.round(users_df['전체페이지수']/users_df['전체세션수'],1)
users_df = users_df.reset_index()

# heatmap
rcParams['figure.figsize'] = 16, 4
col = '세션당페이지수'

print(users_df[['class', col]].groupby('class').describe()) # print(seg별 세션당페이지수 평균)

heatmap_df = users_df[users_df[col]<30]
heatmap_df[col] = heatmap_df[col].apply(np.ceil)
heatmap_df = heatmap_df[['class', col]].groupby(['class', col], as_index=True).size()
heatmap_df = heatmap_df.reset_index()
heatmap_df.columns=['class','세션당페이지수','count']
heatmap_df = heatmap_df.pivot('class', col, 'count').fillna(0)

ax = sns.heatmap(heatmap_df,cmap='viridis', yticklabels=['고이용','미접속자'])
plt.title('Class별 세션당페이지수', fontsize=20)
plt.show() 

# Normalize it by row:
heatmap_df_norm_row = heatmap_df.apply(lambda x: (x-x.mean())/x.std(), axis = 1)
 
# And see the result
sns.heatmap(heatmap_df_norm_row, cmap='viridis', yticklabels=['고이용','미접속자'])
plt.title('Class별 세션당페이지수 (Normalized)', fontsize=20)
plt.show()
```

### 1-2. 카테고리별 UV

```python
# 카테고리별 월별 UV (방문여부)

# 월별 카테고리별 class별 UV수
category_df = df[['카테고리1', 'month', 'class', 'PartyId']].groupby(['month', '카테고리1', 'class'], as_index=True)['PartyId'].nunique().to_frame()
category_df.columns=['순페이지수']

# 카테고리별 월 평균 UV
category_df = category_df.reset_index()
category_df = category_df.groupby(['카테고리1','class'], as_index=True).mean().round(1)
category_df = category_df.reset_index()

# 클래스별 TOP K Visualization 
k = 10
col = '순페이지수'

fig = make_subplots(rows=2, cols=1)
for i in range(1,3):
    cl = i-1
    tmp_df = category_df[category_df['class']==cl]
    rank_df = tmp_df.sort_values(by=[col], ascending=False)[:k]
    rank_df = rank_df.sort_values(by=[col])

    fig.append_trace(go.Bar(
        x=rank_df[col].values,
        y=rank_df['카테고리1'].values,
        orientation='h'
    ), row=i, col=1)

fig.update_layout(height=600, width=1000, title_text="클래스별 UV TOP "+str(k))
fig.show()
```

```python
# 카테고리별 (Normalized) 월별 UV 차이 TOP K
# 비율로
normalized = 'mean' # 'mean' 'min_max'

diff_df = pd.DataFrame()
high, low = category_df[category_df['class']==0]['순페이지수'].values, category_df[category_df['class']==1]['순페이지수'].values
if normalized=='mean':
    high = (high - high.mean()) / high.std()
    low = (low - low.mean()) / low.std()
elif normalized=='min_max':
    high = (high - high.min()) / (high.max()-high.min())
    low = (low - low.min()) / (low.max()-low.min())
diff_df['카테고리1'] = category_df[category_df['class']==0]['카테고리1']
diff_df['고이용-미접속'] = (high - low)
diff_df['미접속-고이용'] = (low - high)

fig = make_subplots(rows=2, cols=1)

col = '고이용-미접속'
rank_df = diff_df[diff_df[col]>0]
rank_df = rank_df.sort_values(by=[col], ascending=False)
rank_df = rank_df.sort_values(by=[col])
fig.append_trace(go.Bar(
    x=rank_df[col].values,
    y=rank_df['카테고리1'].values,
    orientation='h'
), row=1, col=1)

col = '미접속-고이용'
rank_df = diff_df[diff_df[col]>0]
rank_df = rank_df.sort_values(by=[col], ascending=False)
rank_df = rank_df.sort_values(by=[col])
fig.append_trace(go.Bar(
    x=rank_df[col].values,
    y=rank_df['카테고리1'].values,
    orientation='h'
), row=2, col=1)

fig.update_layout(height=450, width=1000, title_text="클래스별 (Normalized) UV 차이 > 0")
fig.show()
```

### 1-3. 카테고리별 종료율

```python
# 카테고리별 종료율

# 페이지 필터링 x
# 1) 생각해보니까 페이지 필터링을 하면 종료 페이지가 달라져서 
# 2) 카테고리별 전체 접속수 / 종료 수 

close = []
sess_id = df['세션ID'].values
for i, now in enumerate(sess_id[:-1]):
    if now == sess_id[i+1]:
        close.append(0)
    else:
        close.append(1)
close.append(1)

df['전체종료여부'] = close

# 종료 수
tmp_df = df[['class', '카테고리1', '전체종료여부']].groupby(['class', '카테고리1'], as_index=True).sum()
tmp_df.columns = ['전체종료수']
df = df.drop(columns = ['전체종료여부'])
gc.collect()

# tmp_df rows -> columns
tmp_df = tmp_df.reset_index()
tmp_df = tmp_df.pivot(index='class', columns='카테고리1', values='전체종료수')
tmp_df = tmp_df.fillna(0)

# 모든 칼럼 / sum
sum_df = tmp_df.sum(axis=1)
tmp_df = tmp_df.apply(lambda x:x/sum_df*100)
close_df = tmp_df.round(2)
close_df
```

```python
# 클래스별 종료율 TOP K Visualization 
k = 10

fig = make_subplots(rows=2, cols=1)
for i in range(1,3):
    cl = i-1
    tmp_df = close_df.iloc[cl]
    rank_df = tmp_df.sort_values(ascending=False)[:k]
    rank_df = rank_df.sort_values()

    fig.append_trace(go.Bar(
        x=rank_df.values,
        y=rank_df.index,
        orientation='h'
    ), row=i, col=1)

fig.update_layout(height=600, width=1000, title_text="클래스별 종료율 TOP "+str(k))
fig.show()
```

```python
# 페이지 종료율 visualization 
# 차이 가장 큰 페이지 TOP 

diff_df = pd.DataFrame()
diff_df['카테고리1'] = close_df.columns
diff_df['고이용-미접속'] = (close_df.loc[0] - close_df.loc[1]).values
diff_df['미접속-고이용'] = (close_df.loc[1] - close_df.loc[0]).values

fig = make_subplots(rows=2, cols=1)

col = '고이용-미접속'
rank_df = diff_df[diff_df[col]>0]
rank_df = rank_df.sort_values(by=[col], ascending=False)
rank_df = rank_df.sort_values(by=[col])
fig.append_trace(go.Bar(
    x=rank_df[col].values,
    y=rank_df['카테고리1'].values,
    orientation='h'
), row=1, col=1)

col = '미접속-고이용'
rank_df = diff_df[diff_df[col]>0]
rank_df = rank_df.sort_values(by=[col], ascending=False)
rank_df = rank_df.sort_values(by=[col])
fig.append_trace(go.Bar(
    x=rank_df[col].values,
    y=rank_df['카테고리1'].values,
    orientation='h'
), row=2, col=1)

fig.update_layout(height=550, width=1000, title_text="클래스별 카테고리 종료률 (고이용-장기)>0 TOP "+str(k))
fig.show()
```

```python
# 세션별 페이지수 (세션 길이)

# groupby class
users_df = df[['PartyId', 'month', 'class']].groupby(['PartyId','month']).first()
users_len = len(users_df)

# 전체 페이지수
tmp_df = df[['PartyId', 'month', '방문일시']].groupby(['PartyId','month']).count()
tmp_df.columns = ['전체페이지수']
users_df = pd.merge(users_df, tmp_df, left_index=True, right_index=True)
assert len(users_df) == users_len, 'Error! length change'

# 전체 세션수
tmp_df = df[['PartyId', 'month','세션ID']].groupby(['PartyId','month'])['세션ID'].nunique().to_frame()
tmp_df.columns = ['전체세션수']
users_df = pd.merge(users_df, tmp_df, left_index=True, right_index=True)
assert len(users_df) == users_len, 'Error! length change'

# 세션별 페이지수
users_df['세션당페이지수'] = np.round(users_df['전체페이지수']/users_df['전체세션수'],1)
users_df = users_df.reset_index()

# heatmap
rcParams['figure.figsize'] = 16, 4
col = '세션당페이지수'

print(users_df[['class', col]].groupby('class').describe()) # print(seg별 세션당페이지수 평균)

heatmap_df = users_df[users_df[col]<30]
heatmap_df[col] = heatmap_df[col].apply(np.ceil)
heatmap_df = heatmap_df[['class', col]].groupby(['class', col], as_index=True).size()
heatmap_df = heatmap_df.reset_index()
heatmap_df.columns=['class','세션당페이지수','count']
heatmap_df = heatmap_df.pivot('class', col, 'count').fillna(0)

ax = sns.heatmap(heatmap_df,cmap='viridis', yticklabels=['고이용','미접속자'])
plt.title('Class별 세션당페이지수', fontsize=20)
plt.show() 

# Normalize it by row:
heatmap_df_norm_row = heatmap_df.apply(lambda x: (x-x.mean())/x.std(), axis = 1)
 
# And see the result
sns.heatmap(heatmap_df_norm_row, cmap='viridis', yticklabels=['고이용','미접속자'])
plt.title('Class별 세션당페이지수 (Normalized)', fontsize=20)
plt.show()
```

### 1-2. 카테고리별 UV

```python
# 카테고리별 월별 UV (방문여부)

# 월별 카테고리별 class별 UV수
category_df = df[['카테고리1', 'month', 'class', 'PartyId']].groupby(['month', '카테고리1', 'class'], as_index=True)['PartyId'].nunique().to_frame()
category_df.columns=['순페이지수']

# 카테고리별 월 평균 UV
category_df = category_df.reset_index()
category_df = category_df.groupby(['카테고리1','class'], as_index=True).mean().round(1)
category_df = category_df.reset_index()

# 클래스별 TOP K Visualization 
k = 10
col = '순페이지수'

fig = make_subplots(rows=2, cols=1)
for i in range(1,3):
    cl = i-1
    tmp_df = category_df[category_df['class']==cl]
    rank_df = tmp_df.sort_values(by=[col], ascending=False)[:k]
    rank_df = rank_df.sort_values(by=[col])

    fig.append_trace(go.Bar(
        x=rank_df[col].values,
        y=rank_df['카테고리1'].values,
        orientation='h'
    ), row=i, col=1)

fig.update_layout(height=600, width=1000, title_text="클래스별 UV TOP "+str(k))
fig.show()
```

```python
# 카테고리별 (Normalized) 월별 UV 차이 TOP K
normalized = 'mean' # 'mean' 'min_max'

diff_df = pd.DataFrame()
high, low = category_df[category_df['class']==0]['순페이지수'].values, category_df[category_df['class']==1]['순페이지수'].values
if normalized=='mean':
    high = (high - high.mean()) / high.std()
    low = (low - low.mean()) / low.std()
elif normalized=='min_max':
    high = (high - high.min()) / (high.max()-high.min())
    low = (low - low.min()) / (low.max()-low.min())
diff_df['카테고리1'] = category_df[category_df['class']==0]['카테고리1']
diff_df['고이용-미접속'] = (high - low)
diff_df['미접속-고이용'] = (low - high)

fig = make_subplots(rows=2, cols=1)

col = '고이용-미접속'
rank_df = diff_df[diff_df[col]>0]
rank_df = rank_df.sort_values(by=[col], ascending=False)
rank_df = rank_df.sort_values(by=[col])
fig.append_trace(go.Bar(
    x=rank_df[col].values,
    y=rank_df['카테고리1'].values,
    orientation='h'
), row=1, col=1)

col = '미접속-고이용'
rank_df = diff_df[diff_df[col]>0]
rank_df = rank_df.sort_values(by=[col], ascending=False)
rank_df = rank_df.sort_values(by=[col])
fig.append_trace(go.Bar(
    x=rank_df[col].values,
    y=rank_df['카테고리1'].values,
    orientation='h'
), row=2, col=1)

fig.update_layout(height=500, width=1000, title_text="클래스별 (Normalized) UV 차이 > 0")
fig.show()
```

### 1-3. 카테고리별 종료율

```python
# 카테고리별 종료율

# 페이지 필터링 x
# 1) 생각해보니까 페이지 필터링을 하면 종료 페이지가 달라져서 
# 2) 카테고리별 전체 접속수 / 종료 수 

close = []
sess_id = df['세션ID'].values
for i, now in enumerate(sess_id[:-1]):
    if now == sess_id[i+1]:
        close.append(0)
    else:
        close.append(1)
close.append(1)

df['전체종료여부'] = close

# 종료 수
tmp_df = df[['class', '카테고리1', '전체종료여부']].groupby(['class', '카테고리1'], as_index=True).sum()
tmp_df.columns = ['전체종료수']
df = df.drop(columns = ['전체종료여부'])
gc.collect()

# tmp_df rows -> columns
tmp_df = tmp_df.reset_index()
tmp_df = tmp_df.pivot(index='class', columns='카테고리1', values='전체종료수')
tmp_df = tmp_df.fillna(0)

# 모든 칼럼 / sum
sum_df = tmp_df.sum(axis=1)
tmp_df = tmp_df.apply(lambda x:x/sum_df*100)
close_df = tmp_df.round(2)
```

```python
# 클래스별 종료율 TOP K Visualization 
k = 10

fig = make_subplots(rows=2, cols=1)
for i in range(1,3):
    cl = i-1
    tmp_df = close_df.iloc[cl]
    rank_df = tmp_df.sort_values(ascending=False)[:k]
    rank_df = rank_df.sort_values()

    fig.append_trace(go.Bar(
        x=rank_df.values,
        y=rank_df.index,
        orientation='h'
    ), row=i, col=1)

fig.update_layout(height=600, width=1000, title_text="클래스별 UV TOP "+str(k))
fig.show()
```

```python
# 페이지 종료율 visualization 
# 차이 가장 큰 페이지 TOP 

diff_df = pd.DataFrame()
diff_df['카테고리1'] = close_df.columns
diff_df['고이용-미접속'] = (close_df.loc[0] - close_df.loc[1]).values
diff_df['미접속-고이용'] = (close_df.loc[1] - close_df.loc[0]).values

fig = make_subplots(rows=2, cols=1)

col = '고이용-미접속'
rank_df = diff_df[diff_df[col]>0]
rank_df = rank_df.sort_values(by=[col], ascending=False)
rank_df = rank_df.sort_values(by=[col])
fig.append_trace(go.Bar(
    x=rank_df[col].values,
    y=rank_df['카테고리1'].values,
    orientation='h'
), row=1, col=1)

col = '미접속-고이용'
rank_df = diff_df[diff_df[col]>0]
rank_df = rank_df.sort_values(by=[col], ascending=False)
rank_df = rank_df.sort_values(by=[col])
fig.append_trace(go.Bar(
    x=rank_df[col].values,
    y=rank_df['카테고리1'].values,
    orientation='h'
), row=2, col=1)

fig.update_layout(height=600, width=1000, title_text="클래스별 카테고리 종료률 (고이용-장기)>0 TOP "+str(k))
fig.show()
```

# 별도 #1 Create new monthly segment_df 

```python
save_df_lst = []

# 2020
for i in range(1, 13): 
    df = load_applog('2020/'+str(i), 2)
    df = basic_preprocess_applog(df)
    
    month_df = df[['PartyId', '방문일시']]
    month_df['month'] = month_df['방문일시'].dt.to_period('M')
    
    # m월 접속자 partyid
    m_df_lst = []
    m_id_lst = []
    m_names = pd.period_range(start='2020/'+str(i), periods=2, freq='M').strftime('%Y-%m')
    for i, m in enumerate(m_names):
        tmp_df = month_df[month_df['month']==m]
        tmp_df = tmp_df[~tmp_df['PartyId'].duplicated(keep='last')]
        m_df_lst.append(tmp_df)
        m_id_lst.append(m_df_lst[i].PartyId.unique())

    # ~(m+1)월: 1, (m+1)월: 0 
    save_df = m_df_lst[0]
    save_df['class'] = [0] * len(save_df) 
    save_df['class'][(~save_df.PartyId.isin(m_id_lst[1]))] = 1
    save_df = save_df.drop(columns=['방문일시'])
    save_df_lst.append(save_df)

# 2021
for i in range(1, 3):
    df = load_applog('2021/'+str(i), 2)
    df = basic_preprocess_applog(df)
    
    month_df = df[['PartyId', '방문일시']]
    month_df['month'] = month_df['방문일시'].dt.to_period('M')
    
    # m월 접속자 partyid
    m_df_lst = []
    m_id_lst = []
    m_names = pd.period_range(start='2021/'+str(i), periods=2, freq='M').strftime('%Y-%m')
    for i, m in enumerate(m_names):
        tmp_df = month_df[month_df['month']==m]
        tmp_df = tmp_df[~tmp_df['PartyId'].duplicated(keep='last')]
        m_df_lst.append(tmp_df)
        m_id_lst.append(m_df_lst[i].PartyId.unique())

    # ~(m+1)월: 1, (m+1)월: 0 
    save_df = m_df_lst[0]
    save_df['class'] = [0] * len(save_df) 
    save_df['class'][(~save_df.PartyId.isin(m_id_lst[1]))] = 1
    save_df = save_df.drop(columns=['방문일시'])
    save_df_lst.append(save_df)

save_df = pd.concat(save_df_lst)
save_df = save_df.reset_index(drop=True)
save_df.to_csv('partyid_monthly_class.csv')
```
